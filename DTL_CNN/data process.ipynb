# Process data to birnary diagnosis

```python
import pandas as pd
import numpy as np

source_train= pd.read_csv('/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/source_train/findings_final_0814_seed1591536269_size10000.csv')
target_test= pd.read_csv("/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/target_test/findings_final_0814_seed-1494714102_size10000.csv")
target_train=pd.read_csv("/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/target_train/findings_final_0814_seed2132231585_size10000.csv")
```

```python
def tobinary(df, diagnosis=None):
    if diagnosis is None:
        return pd.concat([df[df.columns[:-1]], pd.get_dummies(df['diagnosis'])], axis=1)
    else:
        return pd.concat([df[df.columns[:-1]], pd.get_dummies(df['diagnosis'])[diagnosis]], axis=1)
```

```python
tobinary(source_train,'I').to_csv('/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/source_train/findings_final_0814_seed1591536269_size10000_I.csv', index=False)
tobinary(target_test,'I').to_csv('/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/target_test/findings_final_0814_seed-1494714102_size10000_I.csv', index=False)
tobinary(target_train,'I').to_csv('/Users/yuhe/CNN_package/DTL_CNN/data/synthetic_data_v2/target_train/findings_final_0814_seed2132231585_size10000_I.csv', index=False)
```
